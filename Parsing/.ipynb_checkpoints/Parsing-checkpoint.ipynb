```python
reviewsDictionary = {}

i = 0

for review in tmpUserReview:
    try:
        tmpUserName = review.find('span', 'user').text
        tmpDateTime = review.find('span', 'date').text
        tmpRaiting = review.find('div', 'stars-container')['title']
    except:
        print("Одна или более переменных недоступны!")
        tmpUserName = ''
        tmpDateTime = ''
        tmpRaiting = ''
    try:
        review.find('div', 'stars-rating').decompose()
        review.find('div', 'clearfix').decompose()
        review.find('div', 'author').decompose()
    except:
        print("При удалении возникла ошибка!")
    
    try:
        tmpReview = review.text.strip().replace('\r', '').replace('\n', '')
    except:
        print("Возникла ошибка при парсинге комментария!")
    newKey = 'r' + str(i)
    if newKey in reviewsDictionary:
        print(f"Ключ {newKey} уже есть в словаре!")
    else:
        reviewsDictionary[newKey] = {
            'name': tmpUserName,
            'date': tmpDateTime,
            'raiting': tmpRaiting,
            'review': tmpReview
        }
    i += 1
    
#     print(f"Пользователь с никнеймом: {tmpUserName} написал отзыв в {tmpDateTime}, с оценкой: {tmpRaiting}.\nСодержание отзыва: {tmpReview}\n")
```

```python
import csv
with open('commentsWithOtziv.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(('id', 'name', 'date', 'raiting', 'review'))
    i = 1
    for review in reviewsDictionary:
        writer.writerow((i, reviewsDictionary[review]['name'], reviewsDictionary[review]['date'], reviewsDictionary[review]['raiting'], reviewsDictionary[review]['review']))
        i += 1
```
